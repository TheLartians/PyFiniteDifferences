#PyFiniteDifferences - Curved waveguide

We simulate a curved Waveguide with a squared profile using PyFiniteDifferences in 2 and 3 Dimensions.

##Imports
This is where the necessary libraries are imported. Note that we import sympy and the finite difference symbols into the local namespace.

```python
from sympy import *
from pyfinitedifferences.symbols import *
from pyfinitedifferences import finitedifferences
from pyfinitedifferences import settings
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
%matplotlib inline
init_printing(use_latex='mathjax')
```

##Units
This will allow us to use Units in simulation parameters. Here, units are imported into the namespace *u*.

```python
import sympy.physics.units as u
u.keV = 1000 * u.eV
```

##Symbols
Necessary symbols for constants and variables are defined here for later use. Note the use of latex in the variable names for typesetting.

```python
n_Ta = Symbol("n_\mathrm{Ta}") 
n_Va = Symbol("n_\mathrm{Va}")

E    = Symbol("E",real=True,positive=True)
hbar = Symbol(r"\hslash",real=True)
c    = Symbol("c",real=True)

R_waveguide  = Symbol("R_\mathrm{waveguide}",real=True)
R_kruemmung  = Symbol("R_\mathrm{Kruemmung}",real=True)
r_waveguide  = Symbol("r_\mathrm{Waveguide}",real=True,positive=True)
l_waveguide  = Symbol("l_\mathrm{Waveguide}",real=True,positive=True)

alpha = Symbol(r"\alpha")

n_inside     = Symbol("n_\mathrm{Inside}",complex=True)
n_outside    = Symbol("n_\mathrm{Outside}",complex=True)
```

##Settings
The simulation parameters are stored in this object.

```python
setting = settings.AdvancedSettings()
```

##Solve the parabolic wave equation
We will use the [Crank–Nicolson method](http://en.wikipedia.org/wiki/Crank–Nicolson_method) to numerically solve the parabolic wave equation:

$$
\frac{\partial u}{\partial z} = A \left(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2} \right) + F(x,y,z) u, 
$$

with $F(x,y,z)$ and $A$ defined as below.

```python
setting[F] = -I*k/2*(n**2-1)
setting[A] = -I/(2*k)

display.display(Eq(F,setting[F]))
display.display(Eq(A,setting[A]))
```

##Wave number
We will define the the wavenumber $k$ through the wave energy and set the numeric values for $\hbar$ and $c$.

```python
setting.n[hbar] = u.hbar
setting.n[c]    = u.c
setting[k] = E / (hbar*c)

display.display(Eq(k,setting[k]))
```

##Simulation box
The simulation box is defined using $n_x \cdot n_y \cdot n_z$ voxels in $x$, $y$ and $z$ direction (propagation direction is $z$).

```python
setting.numeric_values.update({
        nx:900,
        ny:300,
        nz:600,
        
        sx:0.9*u.um,
        sy:0.3*u.um,
        sz:1200*u.um
})
```

Here we set the unit symbol (for use in in plots) and natural length scale (unit in the numerical simulation) of our simulation by substituting the basic length unit `u.m`. For high numeric precision this should be close to the step size.

```python
setting.length_unit = 'um'
setting.numeric_values[u.m] = u.m/u.um
```

##Refractive index
Here we define the refractive index $n(x,y,z)$ for a curved waveguide with radius of curvature $R_\mathrm{kruemmung}$. The waveguide shape is square with side length $2 \cdot r_\mathrm{waveguide}$. The length of the waveguide in z-direction is $l_\mathrm{waveguide}$. The outside material is chosen to be Ta with index of refraction $n_{Ta}$ and inside the waveguide we assume vacuum.

```python
exprPlus = (R_kruemmung+r_waveguide)
exprMinus = (R_kruemmung-r_waveguide)
expr1 = exprPlus**2 >= (z**2 + (x+R_kruemmung)**2)
expr2 = (z**2 + (x+R_kruemmung)**2) >= exprMinus**2

inside_waveguide = Or(And(expr1, expr2 ,abs(y)<=r_waveguide),z>=l_waveguide) 
n_waveguide = Piecewise((n_inside,inside_waveguide),(n_outside,True))

setting.update({
    n:           n_waveguide,
    n_inside:    n_Va,
    n_outside:   n_Ta,
})
display.display(n_waveguide)
```

##Numeric values
Here numeric values for the simulation are set.

```python
setting.numeric_values.update({
    n_Ta: 1-4.1438E-05-I*3.4516E-06,
    n_Va: 1,
    
    R_kruemmung: 2000*u.mm,
    r_waveguide: 40*u.nm,
    l_waveguide: 800*u.um,
    
    E: 12.5*u.keV,
})
```

##Plot
With numeric values set, we can now plot the simulation box.

```python
plt.figure(figsize=(16,8))
plt.subplot(121)
finitedifferences.plot_front(abs(n),setting,show=False)
plt.subplot(122)
finitedifferences.plot_side(abs(n),setting)
```

##exit angle
And we would like to calculate the exit angle. Don't be confused with the plot above. The angle is much smaller than it seems because of the different sizes of the axises.

```python
alpha = Symbol("alpha_{Austritt}")
setting[alpha] = asin(l_waveguide/R_kruemmung/u.degree)
display.display(Eq(alpha, setting[alpha]))
display.display(Eq(alpha, setting.n[alpha]))
```

The angle is given in $^\circ$

##Boundary conditions
Now boundary conditions are defined. $u_0(x,y)$ is the field at $z=z_0$, while $u_{0\mathrm{Boundary}}(x,y,z)$ is the field at the simulation box edges ($x = \pm s_x/2 \lor y = \pm s_y/2$).

```python
setting[u0] = Number(1)
setting[u0_boundary] = u0 * exp(-I*k*(n_outside-1)*Piecewise((l_waveguide,z>=l_waveguide),(z,True)))
display.display(Eq(u0,setting[u0]))
display.display(Eq(u0_boundary,setting[u0_boundary]))
```

##Finite difference solver 1D
Create the solver using the settings and run a 1D simulation.

```python
solver = finitedifferences.Solver1D(setting)
data = solver.run()
finitedifferences.plot_side(np.abs(data)**2,setting,size=(10,5))
```

##Finite difference solver 2D
Create the solver using the settings and run a 2D simulation. When calling Solver2D.run() be sure to have enough memory available as this will create a 3-Dimensional complex array of size $n_x \cdot n_y \cdot n_z$. Otherwise use the Solver2D.step() and Solver2D.get_field() methods. We integrate the intensity along the y-axis and plot the field.

```python
solver = finitedifferences.Solver2D(setting)
data = solver.run()
finitedifferences.plot_side(np.sum(np.abs(data)**2,axis=1),setting,size=(10,5))
```

Or plot the a cut at a specific $y$-value (which a cut through the data at $y_i = ny/2 + y/dy$).

```python
finitedifferences.plot_side(np.abs(data[:,setting.get_y_index(0),:])**2,setting,size=(10,5))
```

We can also plot the field for a specific $z$-value (which is a cut through the data at $z_i = \frac{z}{dz}$).

```python
finitedifferences.plot_front(np.abs(data[:,:,setting.get_z_index(l_waveguide)])**2,setting,size=(5,5))
```
